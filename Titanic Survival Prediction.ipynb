# Pre-processing

```python
data.columns
```

```python
data.drop('Cabin',axis=1)
```

```python
data=data.drop('Cabin',axis=1)
```

```python
data.columns
```

```python
age_mean=data['Age'].mean()
```

```python
data['Age']=data['Age'].fillna(age_mean)
```

```python
data["Age"].isna().sum()
```

```python
embarked_mode = data["Embarked"].mode()[0]
data["Embarked"] =  data["Embarked"].fillna(embarked_mode)
```

```python
q1 =data['Age'].quantile(0.25)
q3 =data['Age'].quantile(0.75)
iqr = q3-q1

ul=q3+1.5*iqr
Ll=q1-1.5*iqr

data=data[((data.Age>=Ll)&(data.Age<=ul))]
```

```python
q1,q3,iqr,ul,Ll
```

```python
data.shape
```

### removing outliers

```python
age_max = 75
age_mn = 1.5

def cap_age(age):
    if age> age_max:
        return age_max

    if age< age_mn:
        return age_mn
    return age

data['Age'] = data['Age'].apply(lambda x: cap_age (x))

#option
data['Age'] = data['Age'].apply(lambda age: age_max if age> age_max else age)
data['Age'] = data['Age'].apply(lambda age: age_mn if age< age_mn else age)
```

```python
# option 2
def remove_outliers_fare(data):
    q1 =data['Fare'].quantile(0.25)
    q3 =data['Fare'].quantile(0.75)
    iqr = q3-q1

    ul=q3+1.5*iqr
    Ll=q1-1.5*iqr
    data[((data.Fare>=Ll)&(data.Fare<=ul))]
    return data

df_pclass_1 = remove_outliers_fare(data)
df_pclass_2 = remove_outliers_fare(data)
df_pclass_3 = remove_outliers_fare(data)

df_final = pd.concat([df_pclass_1,df_pclass_2,df_pclass_3])
```

## Feature scaling

```python
x_train, x_test, y_train,y_test = train_test_split(features,labels,test_size=0.2,random_state=111222)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(x_train)
x_train=min_max_scaler.transform(x_train)
x_test=min_max_scaler.transform(x_test)
```

```python
#Method 2
min_max_scaler = MinMaxScaler()
#min_max_scaler.fit(x_train)
x_train=min_max_scaler.fit_transform(x_train)
x_test=min_max_scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=4)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

def check_accuracy(y_test,preds):
    print("Accuracy  : {:.2f}%".format(accuracy_score(y_test,preds)*100))
    print("Precision : {:.2f}%".format(precision_score(y_test,preds)*100))
    print("Recall    : {:.2f}%".format(recall_score(y_test,preds)*100))
    print("F1        : {:.2f}%".format(f1_score(y_test,preds)*100))
```

```python
check_accuracy(y_test,predictions)
```

```python
print(f"accuracy {accuracy_score(y_test,predictions)*100:.2f}%")
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
check_accuracy(y_test,predictions)
```
